# 2.	Data Cleansing

```python
df.drop_duplicates()
```

```python
duplicates = df[df['Invoice Date'].duplicated()]
print(duplicates)
```

```python
df['Units Sold'] = df['Units Sold'].str.replace(',', '').astype(float)
```

```python
df.dtypes
```

```python
df.columns
```

```python
df['Product'].unique()
```

```python
df['State'].unique()
```

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'], dayfirst=True)
df['Month'] = df['Invoice Date'].dt.month
df['Year'] = df['Invoice Date'].dt.year
df['Day'] = df['Invoice Date'].dt.day
```

```python
df.dtypes
```

```python
df['Region'].unique()
```

```python
df['Region'] = df['Region'].replace('W', 'West')
df['Region'] = df['Region'].replace('MW', 'Midwest')
df['Region'] = df['Region'].replace('NE', 'Northeast')
df['Region'] = df['Region'].replace('S', 'South')
```
